9. Carry out image sharpening on am image of your choice.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('im03.png', cv2.IMREAD_GRAYSCALE)

denoised = cv2.medianBlur(img, 5)

blurred = cv2.GaussianBlur(denoised, (0, 0), sigmaX=3)

mask = denoised - blurred

sharpened_unsharp = denoised + 1.5 * mask

laplacian_kernel = np.array([[0, -1, 0],
                             [-1, 5, -1],
                             [0, -1, 0]])

sharpened_laplacian = cv2.filter2D(denoised, -1, laplacian_kernel)

# Display results 
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
plt.title('Denoised Original')
plt.imshow(denoised, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title('Unsharp Masking')
plt.imshow(sharpened_unsharp, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title('Laplacian Sharpening')
plt.imshow(sharpened_laplacian, cmap='gray')
plt.axis('off')

plt.show()
```
